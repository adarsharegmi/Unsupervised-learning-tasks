```python
# importing the required libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# reading the csv file

df = pd.read_csv('Data for Q4-5.csv')
```

```python
# listing the unique values for given column

df['NObeyesdad'].unique()
```

There are 7 classes of this column.



```python
# Drop Nobeyesdad with insufficient weight value

df_filtered = df.drop(df[df['NObeyesdad'] == 'Insufficient_Weight'].index)
df_filtered.head()
```

```python
# # grouping 
class1 = ['Normal_Weight', 'Overweight_Level_I', 'Overweight_Level_II']
class2 = [ 'Obesity_Type_I', 'Obesity_Type_II', 'Obesity_Type_III']


df_filtered.replace(to_replace =class1, 
                            value ="class a", inplace=True)
df_filtered.replace(to_replace =class2, 
                            value ="class b", inplace=True)
df_filtered
```

```python
# lets encode the data for gender, family_history_with_overweight, CAEC, smoke, MTRANS
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
x = df_filtered

x['Gender'] = le.fit_transform(x['Gender'])
x.iloc[:, 4] = le.fit_transform(x.iloc[:, 4])
x.iloc[:, 5] = le.fit_transform(x.iloc[:, 5])
x.iloc[:, 8] = le.fit_transform(x.iloc[:, 8])
x.iloc[:, 9] = le.fit_transform(x.iloc[:, 9])
x.iloc[:, 11] = le.fit_transform(x.iloc[:, 11])
x.iloc[:, 14] = le.fit_transform(x.iloc[:, 14])
x.iloc[:, 15] = le.fit_transform(x.iloc[:, 15])
x.iloc[:, 16] = le.fit_transform(x.iloc[:, 16])
x
```

```python
y=x['NObeyesdad']
x = x.iloc[:,:-1]
x
```

```python
print("actual result")
print(y)
```

```python
from sklearn.cluster import KMeans
from sklearn import metrics
model = KMeans(n_clusters=2, init='k-means++', random_state=30)
y_pred = model.fit_predict(x)
```

```python
# finding the accuracy
print("Accuracy:",metrics.rand_score(y, y_pred) *100)

#hence model shows 98% accuracy in test data
from sklearn.metrics import classification_report
print(classification_report(y,y_pred))
```

# using mean shift clustering 
This clustering approach is slightly different than Kmeans approach since the number of clusters will be determined with respect to data rather than declaring by onself. KMeans clustering depends on assumption but in case of mean shift the data points are assigned based upon the data points density.


```python
from sklearn.cluster import MeanShift

ms = MeanShift()
ms.fit(x)
labels = ms.labels_

print(labels)

print(np.unique(labels))
```

```python
# Finding the accuracy
# finding the accuracy
print("Accuracy:",metrics.rand_score(y, labels) *100)

#hence model shows 98% accuracy in test data
from sklearn.metrics import classification_report
print(classification_report(y,labels))
```

# Generalisibility of the model

For generalisibility the classes are hard coded to make labels. Later using the unsupervised techniques the labels are predicted for accuracy. Different evaluation metrics are taken care of such as recall, precision, f1 score.

# About Evaluation
Among different metrics f1 score can be used for better evaluation along with accuracy. The Mean shift seems to perform better and also labels are not required to declared before hand.

# Result
Among these two models as  f1 score and other metrics are better than that of Kmeans clustering, mean shift seems to have better promising result
